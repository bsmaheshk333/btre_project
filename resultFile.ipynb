```python
import pandas as pd
df = pd.read_excel('input_sample.xlsx', sheet_name=5, skiprows=4)
df.rename({"Unnamed: 0":"column0", 'Unnamed: 4': 'column4'}, axis="columns", inplace=True)
df
```

```python
df.drop(['column0'], axis=1, inplace=True) # column0 dropped
df
```

```python
# append task
ap_s1 = pd.read_excel('input_sample.xlsx', sheet_name='DataAppend_1', skiprows=4)
ap_s1
```

```python
ap_s2 = pd.read_excel('input_sample.xlsx', sheet_name='DataAppend_2', skiprows=5)
ap_s2
```

```python
append_result = ap_s1.append(ap_s1, ignore_index=True)
# append Done...

# merge task
# merge table 1
merge_df1 = pd.read_excel('input_sample.xlsx', sheet_name='DataMerge_1', skiprows=2)

# merge table2 
merge_df2 = pd.read_excel('input_sample.xlsx', sheet_name='DataMerge_2', skiprows=11)
merge_df2.rename({'Unnamed: 0': 'col1',}, axis="columns", inplace=True)

# merge_df2.drop(['col1', 'Name'],axis=1, inplace=True) # dropped / commented

merge_res = pd.merge(merge_df1, merge_df2, how='inner')
merge_res.drop(['col1'], axis=1, inplace=True)
merge_res.to_excel('result.xlsx', sheet_name='merge_output')
# Merge Done...
```

```python
# drop duplicate rows
rem_dup_rows = pd.read_excel('input_sample.xlsx', sheet_name='Remove Duplicate rows')
rem_dup_rows.drop_duplicates(subset=['col1', 'col2', 'col3'])
rem_dup_rows
# remove dups done...
```

```python
# ToC extraction
toc = pd.read_excel('input_sample.xlsx', sheet_name='ToC Extraction', skiprows=4)
toc
```

```python
# to have all sheets output in a single xl file
xl_writer = pd.ExcelWriter('hands_on.xlsx',)
append_result.to_excel(xl_writer, sheet_name='append_result')
merge_res.to_excel(xl_writer, sheet_name='merge_result')
rem_dup_rows.to_excel(xl_writer, sheet_name='rem_duplicates_result')
xl_writer.save()
```
